```python
import root_numpy as rnp
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
fname = "test_runs_5files.root"
```

```python
for layer in range(24):
    hit=rnp.root2array(fname, treename="T", branches="hits.hit{}".format(layer))
    plt.hist(hit[hit>-1], bins=np.arange(0,300), 
             label='hit{}'.format(layer), 
             edgecolor='darkblue', color='white',
            linewidth=1.2)
    plt.legend(loc='upper right')
    plt.show()
```
