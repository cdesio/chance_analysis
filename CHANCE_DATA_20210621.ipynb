```python
board = 8
event = 39
fig, ax = plt.subplots(figsize=(8,6))
plt.title('Board {}, event {}'.format(board, event))
x=np.arange(320)
y = test_run.board_data(board)[event]
ax.plot(x,y, c='white')
ax.fill_between(x, y, color='blue')
ax.set_facecolor('k')
ax.set_ylim(0, 4000)
plt.grid()
plt.show()
```

```python

pedestal = np.zeros((31, 320))
for board in range(1, 31):
    if board in test_run.active_boards:
        #for event in range(len(test_run.board_data(board))):
            #if np.any(test_run.board_data(board)[0]>= np.mean(test_run.board_data(board)[event])+200):
        print('board: {}'.format(board))
        pedestal[board]=np.mean(test_run.board_data(board), axis=0)
        plt.plot(np.arange(320), pedestal[board])
        plt.ylim(0,2000)
        plt.show()
```

```python
board = 8
event = 39
fig, ax = plt.subplots(figsize=(8,6))
plt.title('Board {}, event {}'.format(board, event))
x=np.arange(320)
y = test_run.board_data(board)[event] - pedestal[8]
ax.plot(x,y, c='white')
ax.fill_between(x, y, color='blue')
ax.set_facecolor('k')
ax.set_ylim(0, 4000)
plt.grid()
plt.show()
```

```python
for board in range(1, 31):
    if board in test_run.active_boards:
        subtracted_pedestal = np.asarray(list(map(lambda a : a-pedestal[board], test_run.board_data(board))))
        for i, event in enumerate(subtracted_pedestal):
            if np.any(event>=300):
                print('board: {}, event: {}'.format(board, i))
                plt.figure(figsize=(8,6))
                plt.plot(np.arange(320), event)
                plt.ylim(0,2000)
                plt.show()
```
