```python
import numpy as np
import root_numpy as rnp
import uproot3
import matplotlib.pyplot as plt
```

```python
import os
```

```python
file_3p2 = '../data/Run_1k_vertexed-3p2.root'
file_up = '../data/Run_1k_vertexed-constr-up.root'
file_down = '../data/Run_1k_vertexed-constr-down.root'
```

```python
f3p2 = uproot3.open(file_3p2)
vx = f3p2['T'].array('pocascattering')['vx']
vy = f3p2['T'].array('pocascattering')['vy']
vz = f3p2['T'].array('pocascattering')['vz']
```

```python
f_up = uproot3.open(file_up)
vx_up = f_up['T'].array('pocascattering')['vx']
vy_up = f_up['T'].array('pocascattering')['vy']
vz_up = f_up['T'].array('pocascattering')['vz']
```

```python
f_down = uproot3.open(file_down)
vx_down = f_down['T'].array('pocascattering')['vx']
vy_down = f_down['T'].array('pocascattering')['vy']
vz_down = f_down['T'].array('pocascattering')['vz']
```

```python
plt.figure(figsize=(10,7))
plt.title('scattering vertex, z coord: vz')
plt.hist(vz, histtype=u'step', edgecolor='red', linewidth=1.5, bins=50, label='3+2: {} tracks'.format(vz.shape[0]))
plt.hist(vz_up, histtype=u'step', edgecolor='blue', linewidth=1.5, bins=10, label='constr-up: {} tracks'.format(vz_up.shape[0]))
plt.hist(vz_down, histtype=u'step', edgecolor='green', linewidth=1.5, bins=10, label='constr-down: {} tracks'.format(vz_down.shape[0]))
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.hist(vx, histtype=u'step', edgecolor='red', linewidth=1.5, bins=50, label='3+2: {} tracks'.format(vz.shape[0]))
plt.hist(vx_up, histtype=u'step', edgecolor='blue', linewidth=1.5, bins=10, label='constr-up: {} tracks'.format(vz_up.shape[0]))
plt.hist(vx_down, histtype=u'step', edgecolor='green', linewidth=1.5, bins=10, label='constr-down: {} tracks'.format(vz_down.shape[0]))
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.hist(vy, histtype=u'step', edgecolor='red', linewidth=1.5, bins=50, label='3+2: {} tracks'.format(vz.shape[0]))
plt.hist(vy_up, histtype=u'step', edgecolor='blue', linewidth=1.5, bins=10, label='constr-up: {} tracks'.format(vz_up.shape[0]))
plt.hist(vy_down, histtype=u'step', edgecolor='green', linewidth=1.5, bins=10, label='constr-down: {} tracks'.format(vz_down.shape[0]))
plt.legend(loc='upper right')
plt.show()
```
