```python
def find_t(distr_1, distr_2):
    mean_1 = np.mean(distr_1)
    mean_2 = np.mean(distr_2)
    #print(mean_1, mean_2)
    t = np.mean([mean_2, mean_1])
    return t
```

```python
from collections import Counter
```

```python
def count_h2_voxels(hist, ts, bins):
    bin_map = np.digitize(hist.distance_axis(axis=0).distance, bins=bins)
    metric_values = hist.histogram.ravel()
    est_thresholds = np.asarray(ts)[bin_map]
    tot = len(hist.histogram.ravel())
    below_threshold = metric_values<est_thresholds
    over_threshold = metric_values>=est_thresholds

    print('total number of voxels in selection: {}'.format(tot))
    print('distances in : {}'.format(bins[np.unique(bin_map)]))
    print('estimated thresholds: {}'.format(np.asarray(ts)[np.unique(bin_map)]))
    print('\nno. voxels over threshold: {}\nno. voxels under threshold: {}\n'.format(len(np.where(over_threshold)[0]), len(np.where(below_threshold)[0])))
    print('\npercentage v.o.t: {}%'.format(len(np.where(over_threshold)[0])/tot*100))
    print('\n est. volume: {} l'.format(len(np.where(over_threshold)[0])*30.3*30.3*30.3/1000))
    return
```

```python
def efficiency(hist, ts, bins, is_bitumen=False):
    bin_map = np.digitize(hist.distance_axis(axis=0).distance, bins=bins)
    counter = Counter(bin_map)
    est_thresholds = ts[bin_map]
    metric_values = hist.histogram.ravel()
    tot = len(hist.histogram.ravel())
    # Setup selection criterion
    if is_bitumen:
        mask = metric_values < est_thresholds
    else: 
        mask = metric_values >= est_thresholds
    eff_map = np.zeros_like(metric_values)
    eff_map[mask] = 1
    acc = np.sum(eff_map) / tot
    bin_eff = np.zeros_like(bins)
    for i in range(len(bins)):
        if i in counter:
            bin_eff[i] = acc * (counter[i] / tot)
    return bin_eff
```
